# Exploratory Analysis

This notebook provides an early exploratory analysis of the mental health literature dataset used in the DASC 690 capstone project.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/mental_health_literature.csv')
df.head()
```

```python
df.info()
```

```python
df['topic'].value_counts()
```

```python
df['year'].value_counts().sort_index().plot(kind='bar')
plt.title('Publication Year Distribution')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.tight_layout()
plt.show()
```
